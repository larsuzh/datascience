```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sn
```

```python
df_ap_risk_rate = pd.read_csv('project_data/workspace/airPollution_riskfactor_final_rates.csv')
df_ap_risk_rate.head()
```

```python
cols = ['year','co2', 'co2_per_capita', 'Deaths: Outdoor air pollution', 'rate Deaths: Outdoor air pollution']
sn.pairplot(df_ap_risk_rate, vars=cols, hue='region', palette='colorblind')
```

```python
plt.figure(figsize=(8,6))
#fig, ax = plt.subplots(1,3,sharex=True)
# ax[0] = sn.scatterplot(data=df_ap_risk_rate, x='year', y='co2', hue='region')
# ax[1] = sn.scatterplot(data=df_ap_risk_rate, x='year', y='co2_per_capita', hue='region')
# ax[2] = sn.scatterplot(data=df_ap_risk_rate, x='year', y='population', hue='region')
sn.scatterplot(data=df_ap_risk_rate, x='year', y='co2', hue='region')
plt.figure(figsize=(8,6))
sn.scatterplot(data=df_ap_risk_rate, x='year', y='co2_per_capita', hue='region')
plt.figure(figsize=(8,6))
sn.scatterplot(data=df_ap_risk_rate, x='year', y='population', hue='region')
plt.figure(figsize=(8,6))
sn.scatterplot(data=df_ap_risk_rate, x='year', y='Deaths: Outdoor air pollution', hue='region')
plt.figure(figsize=(8,6))
sn.scatterplot(data=df_ap_risk_rate, x='year', y='rate Deaths: Outdoor air pollution', hue='region')
plt.show()
```

```python
plt.figure(figsize=(8,6))
#fig, ax = plt.subplots(1,3,sharex=True)
plt.title('total deaths')
sn.scatterplot(data=df_ap_risk_rate, x='year', y=np.sum(df_ap_risk_rate.iloc[:,6:34],axis=1), hue='region')
plt.figure(figsize=(8,6))
plt.title('total deaths log scale')
sn.scatterplot(data=df_ap_risk_rate, x='year', y=np.log10(np.sum(df_ap_risk_rate.iloc[:,6:34],axis=1)), hue='region')
plt.figure(figsize=(8,6))
plt.title('total death rates')
sn.scatterplot(data=df_ap_risk_rate, x='year', y=np.sum(df_ap_risk_rate.iloc[:,34:],axis=1), hue='region')
plt.figure(figsize=(8,6))
plt.title('total death rates log scale')
sn.scatterplot(data=df_ap_risk_rate, x='year', y=np.log10(np.sum(df_ap_risk_rate.iloc[:,34:],axis=1)), hue='region')
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,10))
fig.suptitle('deaths')
sn.scatterplot(data=df_ap_risk_rate, x='year', y='Deaths: Outdoor air pollution', hue='region',ax=ax1)
sn.scatterplot(data=df_ap_risk_rate, x='year', y=np.sum(df_ap_risk_rate.iloc[:,6:34],axis=1), hue='region',ax=ax2)
plt.title('log co2')
sn.scatterplot(data=df_ap_risk_rate, x='year', y=np.log10(df_ap_risk_rate['co2']), hue='region',ax=ax3)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10), sharey=True)
fig.suptitle('deaths log sca')
plt.title('due to air pollution')
sn.scatterplot(data=df_ap_risk_rate, x='year', y=np.log10(df_ap_risk_rate['Deaths: Outdoor air pollution']), hue='region',ax=ax1)
plt.title('total')
sn.scatterplot(data=df_ap_risk_rate, x='year', y=np.log10(np.sum(df_ap_risk_rate.iloc[:,6:34],axis=1)), hue='region',ax=ax2)
```
