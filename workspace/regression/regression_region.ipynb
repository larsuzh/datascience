```python
# Load all dependencies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import statsmodels.formula.api as sm
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
%matplotlib inline
```

```python
df = pd.read_csv('/Users/larsboesch/Desktop/FS22/DataScience/project/datascience/workspace/regression/co2_pm25_deathrates_cleaned.csv')
```

```python
df["category"] = df["country"].astype("category")
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df
```

```python
df.to_csv('co2_pm25_deathrates_category.csv')
```

```python
w_param = 'category'
x_param = 'co2'
y_param = 'PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)'
z_param = 'rate Deaths - Cause: All causes - Risk: Outdoor air pollution - OWID - Sex: Both - Age: All Ages (Number)'
```

```python

df = pd.DataFrame({'x': df[x_param],
                   'y': df[y_param],
                   'z': df[z_param],
                   'w': df[w_param]
                  })

# linear regression fit
reg = sm.ols(formula='z ~ x + y + w', data=df).fit()
print(reg.summary())
```
