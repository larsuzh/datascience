```python
# Load all dependencies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import statsmodels.formula.api as sm
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
%matplotlib inline
```

# Data cleaning

```python
df = pd.read_csv('co2_pm25_deathrates_cleaned.csv')
df_2 = pd.read_csv('additional_co2_new.csv')
```

```python
df_2["pm_25"] = df_2["PM2.5 air pollution - mean annual exposure (micrograms per cubic meter)"]
df_2["rate_deaths"] = df_2["rate Deaths - Cause: All causes - Risk: Outdoor air pollution - OWID - Sex: Both - Age: All Ages (Number)"]
df_2["Region"] = df["country"].astype("category")
```

```python
df_2 = df_2.drop(['Unnamed: 0', 'Unnamed: 4', 'Unnamed: 9', 'category', "Deaths - Cause: All causes - Risk: Outdoor air pollution - OWID - Sex: Both - Age: All Ages (Number)", "rate Deaths - Cause: All causes - Risk: Outdoor air pollution - OWID - Sex: Both - Age: All Ages (Number)", "PM2.5 air pollution - mean annual exposure (micrograms per cubic meter)"], axis=1)
```

# Regession

```python
w_param = 'category'
x_param = 'co2'
y_param = 'PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)'
z_param = 'rate Deaths - Cause: All causes - Risk: Outdoor air pollution - OWID - Sex: Both - Age: All Ages (Number)'
```

```python
df = pd.DataFrame({'x': df[x_param],
                   'y': df[y_param],
                   'z': df[z_param],
                   'w': df[w_param]
                  })

# linear regression fit
reg = sm.ols(formula='z ~ x + y + w', data=df).fit()
print(reg.summary())
```

```python
reg_2 = sm.ols(formula = "rate_deaths ~ cement_co2_per_capita + coal_co2_per_capita + flaring_co2_per_capita + gas_co2_per_capita + oil_co2_per_capita + other_co2_per_capita + consumption_co2_per_capita + pm_25", data=df_2).fit()
print(reg_2.summary())
```
